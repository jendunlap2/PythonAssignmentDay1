#### Performing Calculations on ints and floats

##### Addition

```python
num1 = 3
num2 = 5.

result = num1 + num2
print(result)
print(type(result))
# int + int = int
# float + float = float
# int + float = float

# short-hand version

result += 2 # result = result(current value) + 2
print(result)
```

##### Subtraction

```python
dif = num2 - num1
print(dif)

# short-hand version 
dif -= 1 # dif = dif - 1
print(dif)

weird_result = 5.4 - 2
print(weird_result)

5.4 == 3.4
```

##### Multiplication

```python
product = num1 * num2
print(product)

# short-hand version
product *= 2 # product = product * 2
print(product)
```

##### Division

```python
quotient = num2 / num1
print(quotient)

# short-hand version
quotient /+ 2 # quotient = quotient / 2
print(quotient)
```

##### Floor Division

```python
result_floor = 100 // 3
print(result_floor)

# short-hand version
result_floor //= 2 # result_floor = result_floor // 2
print(result_floor)
```

##### Modulo

```python
# retuns the remainder of the operation
result_mod = 29 % 5
print(result_mod)

# short-hand version
result_mod %= 3 # result_mod = result_mod % 3
print(result_mod)
```

##### Exponential

```python
square = 5 ** 2
print(square)

# short-hand version
square **= 2 # square = square ** 2
print(square)


# square root
square_root = 625 ** (1/2)
print(square_root)
```

## Exercise #1 <br>
<p>Cube Number Test... Print out all cubed numbers up to the total value 1000. Meaning that if the cubed number is over 1000 break the loop.</p>

```python
for i in range(1000):
    cube = i*i*i
    if cube <= 1000:
        print(cube)
    else:
        break
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didn’t
# An else after a for runs if the for didn’t break

for i in range(1,101):
    if i > 1:
        for num in range(2, i):
            if i % num == 0:
                break
            else:
                print(num)
```
